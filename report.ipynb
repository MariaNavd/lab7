## Теоретическая часть
Данное уравнение второго порядка путем замены $y_1=y$ и $y_2=y'$ сводится к системе двух уравнений первого порядка $$\begin{cases}
y'_1=y_2\\ y'_2=-y_2+4y_1+xe^{-x}\end{cases}$$
Метод Тейлора решает задачу Коши вида $$\begin{cases}
y'(x)=f(x,y(x)), x\in(a,b)\\ y(a)=y^0\end{cases}$$
Приближенное решение в точке $x_{n+1}$ сетки $\omega_N$ строится способом, основанным на разложении решения в ряд Тейлора в предыдущей точке сетки $x_n$ по степеням шага $h$: $$y(x_{n+1})=y(x_n)+h\triangle(x_n,y_n,h),$$$$\triangle(x,y,h)\equiv y'(x)+\frac{h}{2}y''(x)+\frac{h^2}{3!}y'''(x)+...$$
Взяв вместо этого ряда конечный его отрезок $$\varphi_p(x,y,h)\equiv y'(x)+\frac{h}{2}y''(x)+...+\frac{h^{p-1}}{p!}y^{(p)}(x)$$ и заменяя в нем производные $y^{(k)}(x)$ $$y'(x)=f(x,y(x)),$$ $$y''(x)=\frac{d}{dx}f(x,y(x))=f'_x+f'_yf,$$ $$y'''(x)=\frac{d}{dx}(f'_x+f'_yf) = f''_{xx}+f'_xf'_y+((f'_y)^2+2f''{xy})f+f''_{yy}f^2,...,$$ получаем последовательность приближений $$y(x_{n+1})=y(x_n)+h\varphi_p(x_n,y_n,h), n=0,1,2,...$$
Для системы из $s$ уравнений первого порядка, где $$y(x)\equiv[y^1(x),...,y^s(x)],$$ $$f(x,y)\equiv[f^1(x,y),...,f^s(x,y)],$$ метод Тейлора можно применить покомпонентно. Тогда при $p=3$ в векторной записи последовательность приближений примет вид $$y(x_{n+1})= y(x_n)+h(f(x_n,y_n)+\frac{h}{2}f'(x_n,y_n)+\frac{h^2}{6}f''(x)),$$ где теперь $$f'(x,y)\equiv\frac{\delta f}{∂x}(x,y)+J(x,y)f(x,y)$$ $$f''(x,y)\equiv\frac{\delta^2f}{\delta x^2}+\frac{\delta f}{\delta x}J(x,y)+((J(x,y))^2+2\frac{\delta J(x,y)}{\delta x})f+S(x,y)f^2$$ $J(x,y)\equiv(\delta f^i/\delta y^j)$ - матрица Якоби функции $f$, $S(x,y)\equiv(\delta^2f^i/(\delta y^i)^2)$
